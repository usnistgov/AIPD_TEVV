```python
plate_start = 'A1' # first well to start at if reading whole plate or continuous row / rows of a plate (no gaps from one row to the next)

plate_type  = '384'
partial_plate = 0 # 0 if the full plate was read, otherwise 1

delim = ',' # delimiter for parsing text file
#delim = None # if tab separated


sNames = ['0','o','1','2','13','14','22','23','49','98'] # names of samples (optional)

rep1 = ['A1','B1','C1','D1',
        'D2','D3','D4']
rep2 = []
rep3 = []

plate_positions = rep1 + rep2 + rep3

nSamples = len(rep1)
nReps = int(len(plate_positions)/nSamples)


nChannels = ['OD600','OD700'] # names of the channels being measured / imported (optional)
# header in text file to start pulling data from
skip_delim = ['Read 1:600','Read 1:700']
# next header in text file to stop pulling data
row_stop_delim = ['Read 1:700','Results']

# Normalize reporter data (0) or (1)
use_norm = 0
# If set to 1 include which channel position has the reporter in it
rep_pos = [0,1] # 0 = first channel, 1 = second channel, etc...
```

```python
chan_names_dash = []
for n in range(len(skip_delim)):
    check = 0
    for m in range(len(skip_delim[n])):
        if skip_delim[n][m] == ':':
            chan_names_dash.append(skip_delim[n][0:m]+'-'+skip_delim[n][m+1:])
            check = 1
    if check == 0:
        chan_names_dash.append(skip_delim[n])
        
        
        
data_dict = len(skip_delim)*[0]
DDn_dict = len(skip_delim)*[0]
DDn_mean = len(skip_delim)*[0]
DDn_std = len(skip_delim)*[0]
use_cols = []
t = []

well_hash = {}
if plate_type == '96':
    wrow = ['A','B','C','D','E','F','G','H']
    tot_cols = 12
elif plate_type == '384':
    wrow = ['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P']
    tot_cols = 24
else:
    print('Specify the plate type')
wcol = [i+1 for i in range(tot_cols)]
for n in range(len(wrow)):
    for m in range(len(wcol)):
        well_hash[wrow[n]+str(wcol[m])] = m+tot_cols*n+2


# creates a dictionary so you can pull data out by well name
def create_data_dict(plate_positions,data,drange=-1):
    
    data_dict = {}
    
    for n in range(len(plate_positions)):
        if drange == -1:
            data_dict[plate_positions[n]] = data[:,n]
        else:
            data_dict[plate_positions[n]] = data[0:drange+1,n]
    
    return data_dict


for j in range(len(skip_delim)):
    
    # opening file and finding where the data starts
    # doing this so that np.loadtxt can directly pull in all the numeric data for you
    skip_to = 0
    with open(results_path+file_path,'r') as handle:
        i = 0
        for line in handle:
            if skip_delim[j] in line:
                print(line)
                skip_to = i+3
            elif ('0:00:00			' in line or '0:00:00	0.0		' in line) and skip_to != 0:
                row_stop = i-1
                break
            elif row_stop_delim[j] in line:
                row_stop = i-2
                break
            i += 1
    
    if plate_positions == 'all':
        plate_postitions = [well_hash[plate_positions[i]] for i in range(len(plate_positions))]
        use_cols = [well_hash[plate_positions[i]] for i in range(len(plate_positions))]        
    elif partial_plate != 0: 
        start_well = 2
        use_cols = [start_well+i for i in range(len(plate_positions))]
    else:
        start_well = well_hash[plate_start]
        use_cols = [well_hash[plate_positions[i]] for i in range(len(plate_positions))] 
    
    
    # import data as numpy array
    data = np.loadtxt(results_path+file_path,skiprows=skip_to,max_rows=row_stop-skip_to,usecols=use_cols,delimiter=delim)
    # import time as string
    t_data = np.loadtxt(results_path+file_path,dtype='str',skiprows=skip_to,max_rows=row_stop-skip_to,usecols=[0],delimiter=delim)
    
    t_sec = np.zeros([len(t_data),1])
    
    # converting time data into seconds and adding to numpy array
    for n in range(len(t_data)):
        if len(t_data[n]) == 7:
            hr = int(t_data[n][0:1])*3600
            mn = int(t_data[n][2:4])*60
            t_sec[n] = hr+mn+int(t_data[n][5:7])
        else:
            hr = int(t_data[n][0:2])*3600
            mn = int(t_data[n][3:5])*60
            t_sec[n] = hr+mn+int(t_data[n][6:8])
            
    
    '''
    ##############################################################################
    USER DEFINED NORMALIZATION OF DATA OF INTEREST
    ##############################################################################
    '''
    
    # pulling out the data you want to plot
    data_dict[j] = create_data_dict(plate_positions,data)
    DDn = {}
    
    # for normalizing the data before plotting
    for n in range(len(plate_positions)):
        
        D = data_dict[j][plate_positions[n]]
        
        if use_norm == 1 and j == rep_pos[j]:
            DDn[plate_positions[n]] = (data_dict[j][plate_positions[n]] - min(data_dict[j][plate_positions[n]]))/ (max(data_dict[j][plate_positions[n]]) - min(data_dict[j][plate_positions[n]]))
        else:
            DDn[plate_positions[n]] = D
    
    DDn_dict[j] = DDn
    
    # converting time to hours
    t.append(t_sec[0:len(D)]/3600)
    
    
    #######################################
    # Computing mean and standard deviation
    #######################################
        
    DDn_mean0 = {}
    DDn_std0 = {}
    
    # for normalizing the data before plotting
    for n in range(len(rep1)):
        reps_for_mean = []
        for m in range(nReps):
            
            reps_for_mean.append(DDn_dict[j][plate_positions[n+m*nSamples]])
        
        DDn_mean0[rep1[n]] = np.mean(reps_for_mean,axis=0)
        DDn_std0[rep1[n]] = np.std(reps_for_mean,axis=0)
    
    DDn_mean[j] = DDn_mean0
    DDn_std[j] = DDn_std0
        
```

```python
### SAVING ALL KINETIC DATA TO EXCEL FILE


###################################
# Saving all individual replicates
###################################
save_loc = results_path+folder_path # saviing in same folder the text file data was pulled from
save_name = file_name[:-4] # same name as the text file name

wb = opxl.Workbook() # to create a new workbook
#wb = opxl.load_workbook(save_loc+save_name+'.xlsx') # to append to an existing workbook

for j in range(len(skip_delim)):
    
    i = len(skip_delim)-(j+1) # creating sheets in the same order they are specified above
    
    ws1 = wb.create_sheet(chan_names_dash[i],0) # giving each sheet name the names of the channels
    
    ws1.cell(1,1,file_name) # saving raw data file
    ws1.cell(2,1,'Raw data') # saving raw data file
    ws1.cell(3,1,'time (min)') # saving raw data file
    
    # saving time data
    for m in range(len(t[i])):
        ws1.cell(m+4,1,t[i][m,0]*60) 
    
    # saving raw data
    for n in range(len(plate_positions)):
        ws1.cell(3,n+2,plate_positions[n])
        for p in range(len(t[i])):
            ws1.cell(p+4,n+2,data_dict[i][plate_positions[n]][p]) 
    
    # saving normalized data if applicable
    if use_norm == 1:
        for n in range(len(plate_positions),2*len(plate_positions)):
            ws1.cell(2,len(plate_positions)+3,'Normalized data')
            ws1.cell(3,n+3,plate_positions[n-len(plate_positions)])
            for p in range(len(t[i])):
                ws1.cell(p+4,n+3,DDn_dict[i][plate_positions[n-len(plate_positions)]][p])         
    
wb.save(save_loc+save_name+'.xlsx')
```

```python
#####################################
# Saving mean and std of each sample 
#####################################
if rep2 != []:
    save_loc = results_path+folder_path # saviing in same folder the text file data was pulled from
    save_name = file_name[:-4]+' means' # same name as the text file name
    
    wb = opxl.Workbook() # to create a new workbook
    #wb = opxl.load_workbook(save_loc+save_name+'.xlsx') # to append to an existing workbook

    for j in range(len(skip_delim)):
        
        i = len(skip_delim)-(j+1) # creating sheets in the same order they are specified above
        
        ws1 = wb.create_sheet(chan_names_dash[i],0) # giving each sheet name the names of the channels
        
        ws1.cell(1,1,file_name) # saving raw data file
        ws1.cell(2,1,'Raw data') # saving raw data file
        ws1.cell(3,1,'time (min)') # saving raw data file
        
        # saving time data
        for m in range(len(t[i])):
            ws1.cell(m+4,1,t[i][m,0]*60) 
        
        # saving raw data
        for n in range(len(rep1)):
            ws1.cell(3,2*n+2,rep1[n]+' ave')
            ws1.cell(3,2*n+3,rep1[n]+' std')
            for p in range(len(t[i])):
                ws1.cell(p+4,2*n+2,DDn_mean[i][rep1[n]][p]) 
                ws1.cell(p+4,2*n+3,DDn_std[i][rep1[n]][p]) 
                
    wb.save(save_loc+save_name+'.xlsx')
    
```

```python
###############################################################################
BAR PLOTS AT A SPECIFIC TIME POINT
###############################################################################


tpt = 47*60 # time (min) to take fluorescence value for bar plot
tptLoc = np.where(np.round(t[0]*60,-1)==tpt)[0][0]


# names of samples (these are now needed for the x-axis of the bar plot)
sNames = ['A1','B1','C1','D1','D3','D4','D5'] 

fs = 12
scale = ['linear','linear'] # log or linear scale, each entry in this list is for the different channels
width = 0.5 # bar width
ymax = [5,5] # each entry in this list is for the different channels
ymin = [0,0] # each entry in this list is for the different channels
alp = 0.4 # bar transparency
mrks = 5 # scatter plot dot size
CPB = ['blue','green'] # colors for each channel of the bar plot

###############################################################################
# making dataframe to store data
bar_data = pd.DataFrame()
bar_data['Sample'] = nReps*sNames

# looping through channels
for m in range(len(skip_delim)):
    bar_data_temp = []
    # looping through replicates
    for v in range(nReps):
        # looping through samples
        for n in range(len(rep1)):
        
            # storing data for first, second, third, etc reps of each sample
            bar_data_temp.append(DDn_dict[m][plate_positions[n+v*nSamples]][tptLoc])
    
    # Saving the channel data in a Pandas array
    bar_data[nChannels[m]] = bar_data_temp
```
